```python
# Step 1: Upload Image
from google.colab import files
uploaded = files.upload()

# Step 2: Import Libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Get uploaded image name
image_name = list(uploaded.keys())[0]

# Step 3: Read Image
img = cv2.imread(image_name)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 4: Region Growing Function
def region_growing(image, seed, threshold=10):
    height, width = image.shape
    segmented = np.zeros((height, width), np.uint8)

    seed_value = image[seed]
    stack = [seed]

    while stack:
        x, y = stack.pop()

        if segmented[x, y] == 0:
            if abs(int(image[x, y]) - int(seed_value)) < threshold:
                segmented[x, y] = 255

                # Check 4 neighbors
                if x > 0:
                    stack.append((x-1, y))
                if x < height-1:
                    stack.append((x+1, y))
                if y > 0:
                    stack.append((x, y-1))
                if y < width-1:
                    stack.append((x, y+1))

    return segmented

# Step 5: Select Seed Point (center of image)
h, w = gray.shape
seed_point = (h//2, w//2)

# Step 6: Apply Region Growing
segmented_image = region_growing(gray, seed_point, threshold=15)

# Step 7: Show Results
print("Original Image:")
cv2_imshow(img)

print("Grayscale Image:")
cv2_imshow(gray)

print("Segmented Region:")
cv2_imshow(segmented_image)
```
