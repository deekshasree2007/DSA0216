```python
# Install OpenCV (if not installed)
!pip install opencv-python

import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

# Upload image
uploaded = files.upload()

# Read the uploaded image
image = cv2.imread(list(uploaded.keys())[0])

# Check if image loaded
if image is None:
    print("Error loading image")
else:
    # Get dimensions
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)

    # Clockwise rotation (45 degrees)
    rotation_matrix_cw = cv2.getRotationMatrix2D(center, -45, 1.0)
    rotated_cw = cv2.warpAffine(image, rotation_matrix_cw, (w, h))

    # Counter-Clockwise rotation (45 degrees)
    rotation_matrix_ccw = cv2.getRotationMatrix2D(center, 45, 1.0)
    rotated_ccw = cv2.warpAffine(image, rotation_matrix_ccw, (w, h))

    # Display images
    print("Original Image")
    cv2_imshow(image)

    print("Clockwise Rotation (45 degrees)")
    cv2_imshow(rotated_cw)

    print("Counter-Clockwise Rotation (45 degrees)")
    cv2_imshow(rotated_ccw)
```
