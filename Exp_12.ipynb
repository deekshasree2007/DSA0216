```python
# Install OpenCV (if needed)
!pip install opencv-python

import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

# Upload image
uploaded = files.upload()

# Read image
image = cv2.imread(list(uploaded.keys())[0])

if image is None:
    print("Error loading image")
else:
    rows, cols = image.shape[:2]

    # Select 4 points from original image
    pts1 = np.float32([[50, 50],
                       [cols-50, 50],
                       [50, rows-50],
                       [cols-50, rows-50]])

    # Define corresponding 4 points in output image
    pts2 = np.float32([[0, 0],
                       [cols, 0],
                       [100, rows],
                       [cols-100, rows]])

    # Get Perspective Transformation matrix
    matrix = cv2.getPerspectiveTransform(pts1, pts2)

    # Apply Perspective Transformation
    transformed_image = cv2.warpPerspective(image, matrix, (cols, rows))

    # Display images
    print("Original Image")
    cv2_imshow(image)

    print("Perspective Transformed Image")
    cv2_imshow(transformed_image)
```
