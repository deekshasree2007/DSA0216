```python
# Install OpenCV (if needed)
!pip install opencv-python

import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

# Upload image
uploaded = files.upload()

# Read image
image = cv2.imread(list(uploaded.keys())[0])

if image is None:
    print("Error loading image")
else:
    # Get dimensions
    (h, w) = image.shape[:2]

    # Define translation distances
    tx = 100   # Move right by 100 pixels
    ty = 50    # Move down by 50 pixels

    # Create translation matrix
    translation_matrix = np.float32([[1, 0, tx],
                                     [0, 1, ty]])

    # Apply translation
    moved_image = cv2.warpAffine(image, translation_matrix, (w, h))

    # Display images
    print("Original Image")
    cv2_imshow(image)

    print("Moved Image (Right 100px, Down 50px)")
    cv2_imshow(moved_image)
```
