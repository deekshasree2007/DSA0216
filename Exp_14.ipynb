```python
# Install OpenCV (if needed)
!pip install opencv-python

import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

# Upload image
uploaded = files.upload()

# Read image
image = cv2.imread(list(uploaded.keys())[0])

if image is None:
    print("Error loading image")
else:
    rows, cols = image.shape[:2]

    # Source points (4 points in original image)
    src_pts = np.float32([[50, 50],
                          [cols-50, 50],
                          [50, rows-50],
                          [cols-50, rows-50]])

    # Destination points (new perspective view)
    dst_pts = np.float32([[0, 0],
                          [cols, 0],
                          [100, rows],
                          [cols-100, rows]])

    # Compute Homography matrix
    H, status = cv2.findHomography(src_pts, dst_pts)

    # Apply Homography transformation
    transformed_image = cv2.warpPerspective(image, H, (cols, rows))

    # Display results
    print("Original Image")
    cv2_imshow(image)

    print("Homography Transformed Image")
    cv2_imshow(transformed_image)
```
