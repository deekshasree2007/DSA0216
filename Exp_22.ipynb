```python
# Step 1: Import Libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

# Step 2: Upload Image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Step 3: Read Image
img = cv2.imread(image_path)

# Step 4: Convert to Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 5: Define Laplacian Kernel (Positive Center)
kernel = np.array([[0, -1, 0],
                   [-1, 4, -1],
                   [0, -1, 0]])

# Step 6: Apply Laplacian Filter
laplacian = cv2.filter2D(gray, -1, kernel)

# Step 7: Sharpen Image (Original + Laplacian)
sharpened = cv2.add(gray, laplacian)

# Step 8: Display Results
print("Original Image")
cv2_imshow(gray)

print("Laplacian Image")
cv2_imshow(laplacian)

print("Sharpened Image")
cv2_imshow(sharpened)
```
